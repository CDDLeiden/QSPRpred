# Searching, Filtering and Plotting Data Sets

```python
import pandas as pd

from qsprpred.data import QSPRDataset

df = pd.read_csv('../../tutorial_data/A2A_LIGANDS.tsv', sep='\t')
dataset = QSPRDataset(
    df=df,
    store_dir="tutorial_output/data",
    name="QuickStartDataset",
    target_props=[{"name": "pchembl_value_Mean", "task": "REGRESSION"}],
    random_state=42
)
dataset.getDF()
```

```python
query_result = dataset.searchWithSMARTS(
    ["n1ncnc1", "[ar]1[ar][n][ar][ar][n]1", "c1cocc1"],
    operator="and")
query_result.getDF()
```

```python
from qsprpred.plotting.grid_visualizers import table_to_grid

table_to_grid(query_result)
```

```python
query_result = dataset.searchWithSMARTS(
    ["n1ncnc1", "[ar]1[ar][n][ar][ar][n]1", "c1cocc1"],
    operator="or")
table_to_grid(query_result)
```

```python
query_result = dataset.searchWithSMARTS(
    ["n1ncnc1", "[ar]1[ar][n][ar][ar][n]1"],
    operator="or").searchWithSMARTS(["c1cocc1"])
table_to_grid(query_result)
```
